We need to calculate the mean house price for each year per borough, so we need to group by the London_Borough and Year columns and apply the mean at the end. 

```python
dfg = df.groupby(by=['London_Borough', 'Year']).mean()
dfg.sample(10)
```

```python
dfg = dfg.reset_index() # reset the index for the new DataFrame dfg
dfg.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 2001 and 2021. **(I later changed this to 1998 to 2018)**
4. Calculate the price ratio as : Average Price for 2021 divided by the Average Price for 2001.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 2001 and 2021 **(I later changed this to 1998 to 2018)**

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham']) # test price ratio function 
```

We want to do this for all of the London Boroughs.

```python
final = {}
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough'] == b]
    final[b] = create_price_ratio(borough)

print(final) 
```

Now we have a dictionary with data about the ratio of average prices for each borough between 2001 and 2021,  we can make it prettier by converting it to a DataFrame. 

```python
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
# We need to transpose it, and reset the index! 
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
df_ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_ratios.head()
```

Let's sort in descending order and select the top 15 boroughs.

```python
top15 = df_ratios.sort_values(by='2018',ascending=False).head(15)
print(top15)
```

Let's plot the boroughs that have seen the greatest changes in price

```python
ax = top15[['Borough','2018']].plot(kind='bar', title = "Ratio of house prices (2018 to 1998) for top 15 London's Borough over the last two decades")

ax.set_xticklabels(top15.Borough)
```
